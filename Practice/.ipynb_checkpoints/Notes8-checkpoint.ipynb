## Sampling

#### Acceptance-Rejection Method

##### Generate for pdf = 2y , 0<y<1

```r
n       = 10000
counter = 0
y       = rep(NA,n)
while(counter<n){
    x <- runif(1)
    u <- runif(1)
    if(u<x){
        counter    = counter +1
        y[counter] = x
    }
}
```

##### Generate Beta(2,2)

```r
n       = 10000
counter = 0
y       = rep(NA,n)
while(counter<n){
    x <- runif(1)
    u <- runif(1)
    if(u<x*(1-x)){
        counter    = counter + 1
        y[counter] = x
    }
}
```
