## Monte Carlo for Inference

Example - Test the height of the group to be 160.  
$H_0: \mu = 160$  
$H_1: \mu \neq 160$

if x_i are normally distributed, under H_0 the appropriate test statistics is  
$T = \frac{\bar{X}-\mu}{\frac{S}{\sqrt{n}}}$

What if distributional form of T in which case we do not know the value of c.

```r
dat = read.table('http://www.ams.stonybrook.edu/~pfkuan/Teaching/AMS597/Data/HeightData.txt',header=T)
qqnorm(dat$height)
qqline(dat$height)
```

```r
n = dim(dat)[1]
m = 10000
mu.est = mean(dat$height)
sd.est = sd(dat$height)
T.obs  = (mu.est-160)*sqrt(n)/sd.est
T.rand = rep(NA,m)
for(i in 1:m){
    x = rnorm(n,160,sd.est)
    T.rand[i] = (mean(x)-160)/sd(x)
}
```

```r
pvalue.est = length(which(abs(T.rand)>=abs(T.obs)))/m
```

## Power Estimation

```r
cL = quantile(T.rand,0.05)
cH = quantile(T.rand,0.95)
n  = 20
m  = 1000
mu_range = seq(161,170)
beta = rep(NA,length(mu_range))
for(i in 1:length(mu_range)){
    T.rand.pwr = rep(NA,m)
    for(j in 1:m){
        x <- rnorm(20,mu_range[i],sd.est)
        T.rand.pwr[j] = (mean(x)-160)*sqrt(n)/sd(x)
    }
    beta[i] = length(which(T.rand.pwr>cL&T.rand.pwr<cH))/m
}
```

```r
plot(mu_range,1-beta)
```
